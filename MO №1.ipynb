### **1.1 Вероятностное пространство, формула Байеса**

**Задание №1. Определить (с обоснованием), зависимы или независимы следующие события:**

(a) Несовместные события;

(b) События, образующие σ-алгебру Σ в пространстве (Ω, Σ, P);

(c) События, имеющие одинаковую вероятность?

**Ответ:**

Два события A и B называют *независимыми*, если появление одного из них никак не влияет на появление друого, то есть условная вероятность А в преположении, что В произошло, совершенно такая же, как и без этого предположения:

$$
\mathrm{P(A|B)} \;=\; {P(A)}
$$

В противном случае события А и В называют *зависимыми*.

(а) События A1, . . . , An называются несовместными, если они взаимо исключают друг друга, то есть никакие два из них не могут появиться вместе.

Если события А и В несовместны, тогда независимыми они будут только в том случае, если P(A) = 0 или P(B) = 0, то есть Р(А ∩ В) = 0. Но для независимых событий должно выполнять условие Р(А ∩ В) = Р(А)Р(В). Это означает, что в невырожденном случае (когда вероятности событий положительны) **несовместные события не могут быть независимыми.**

(b) Множество Σ, элементами которого являются подмножества множества Ω (не обязательно все) называется σ-алгеброй. σ-алгебра содержит как зависимые, так и независимые события.

Например, пусть Ω (пространство элементарных исходов) - это все возможные результаты двух подбрасовыний монеты.Тогда σ-алгебра будет выгледеть следующим образом Σ = {0, {ОР}, {РО}, {ОО}, {РР}, {ОР, ОО}, ..., {OO, OP, PO, PP}}. События "Выпадение орла при первом броске" и "Выпадение решки при втором броске" являются независимыми, так как выполняется условие Р(А ∩ В) = Р(А)Р(В) ($ \frac{1}{4} =  \frac{1}{2} × \frac{1}{2} $). При этом собятия "Выпадение орла при первом броске" и "Выпадение хотя бы одного орла" уже являются зависимыми, так как условие Р(А ∩ В) = Р(А)Р(В) не выполняется ($ \frac{1}{2} \neq \frac{1}{2} × \frac{3}{4} $).

Таким образом, события, образующие σ-алгебру Σ в пространстве (Ω, Σ, P) могут быть **как зависимыми, так и независимыми.**

(с) События, имеющие одинаковую вероятность, **могут быть и зависимыми, и независимыми**. Например, при игре в игральные кости вероятность выпадение 1, 2, 3, 4, 5, 6 - одинковая и равна 1/6, в данной ситуации события являются несовместными, а значит зависимыми (исходя из логике, описанной в пункте (а)). Но если, например, событие А - на первом игральном кубике выпало число меньше 4 (1,2,3), а событие B - на втором игральном кубике выпало нечетное число, то вероятность у этих событий будет одинаковая - 0.5, но при этом события независимые.

### **1.2 Случайный вектор и числовые характеристики**

**Задание №1. Пусть**
$$ f_{\xi}(x, y) = \frac{e^{-2|y|}}{\pi(1 + x^2)} $$

**Является ли данная функция плотностью распределения случайного вектора?**

Плотность распределения должна обладать следующими свойствами:
1. $ f_{\xi}(x, y) \geq 0 $ для всех $ x $ и $ y $.
2. $ \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{\xi}(x, y) \, dx \, dy = 1 $.

Проверим, удовлетворяет ли функция $ f_{\xi}(x, y) = \frac{e^{-2|y|}}{\pi(1 + x^2)} $ данным условиям.

1. $ e^{-2|y|} > 0$ и $ \pi(1 + x^2) > 0$ - условие неотрицательности выполняется, так как и числитель, и знаменатель всегда положительны для всех $ x $ и $ y $.
2. $ \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \frac{e^{-2|y|}}{\pi(1 + x^2)} \, dx \, dy = \int_{-\infty}^{\infty} \frac{dx}{\pi(1 + x^2)}\ × \int_{-\infty}^{\infty} e^{-2|y|}dy \ = \frac{1}{\pi}\ \int_{-\infty}^{\infty} \frac{dx}{1 + x^2}\ × \int_{-\infty}^{\infty} e^{-2|y|}dy \ $

Рассчитаем отдельно интеграл по $ x $ и интеграл по $ y $.

Интеграл по $ x $:

$ \int_{-\infty}^{\infty} \frac{1}{1 + x^2} \, dx = \left[ \arctan(x) \right]_{-\infty}^{\infty} = \frac{\pi}{2} - \left( -\frac{\pi}{2} \right) = \pi $

Интеграл по $ y $:
$ \int_{-\infty}^{\infty} e^{-2|y|} \, dy = 2 \int_{0}^{\infty} e^{-2y} \, dy = 2 \left[ -\frac{1}{2} e^{-2y} \right]_{0}^{\infty} = 2 \cdot \frac{1}{2} = 1 $

Посчитаем полученные значения:

$\frac{1}{\pi}\ \int_{-\infty}^{\infty} \frac{dx}{1 + x^2}\ × \int_{-\infty}^{\infty} e^{-2|y|}dy \ = \frac{1}{\pi}\ × \pi × 1 = 1$

**Ответ:**
$f_{\xi}(x, y) = \frac{e^{-2|y|}}{\pi(1 + x^2)} $ является  плотностью распределения случайного вектора

**Задание №2. Совместное распределение случайных величин ξ и η задано следующей
таблицей:**

| $ \xi \backslash \eta $ | -1 | 0 | 1 |
|-------------------------|----|---|---|
| -1 | $\frac{1}{8}$ | $\frac{1}{12}$ | $\frac{7}{24}$ |
| 1 | $\frac{1}{3}$ |  $\frac{1}{6}$| 0 |

 (a) Найти маргинальные распределения $\xi$ и $\eta$

 (b) Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(\xi, \eta)$

 (c) Исследовать $\xi$ и $\eta$ на независимость и некоррелированность

**Ответ:**

**(a) Маргинальные распределения $\xi$ и $\eta$**

Для $\xi$:

$ P(\xi = -1) = \frac{1}{8} + \frac{1}{12} + \frac{7}{24} = \frac{3}{24} + \frac{2}{24} + \frac{7}{24} = \frac{12}{24} = \frac{1}{2} $

$ P(\xi = 1) = \frac{1}{3} + \frac{1}{6} + {0} = \frac{2}{6} + \frac{1}{6} = \frac{3}{6} = \frac{1}{2} $

Для $\eta$:

$ P(\eta = -1) = \frac{1}{8} + \frac{1}{3} = \frac{3}{24} + \frac{8}{24} = \frac{11}{24} $

$ P(\eta = 0) = \frac{1}{12} + \frac{1}{6} = \frac{1}{12} + \frac{2}{12} = \frac{3}{12} = \frac{1}{4} $

$ P(\eta = 1) = \frac{7}{24} + {0} = \frac{7}{24} $

**(b) Математическое ожидание, ковариационная и корреляционная матрицы вектора $(\xi, \eta)$**

1.Математическое ожидание:

$\mathsf{E}[\xi] = (-1) × \frac{1}{2}\ + 1 × \frac{1}{2} = 0 $

$\mathsf{E}[\eta] = (-1) × \frac{11}{24}\ + 0 × \frac{1}{4} + 1 × \frac{7}{24}= -\frac{4}{24} = -\frac{1}{6} $

$\mathsf{E}[\xi\eta] = (-1) × (-1) × \frac{1}{8}\ + (-1) × 0 × \frac{1}{12} + (-1) × 1 × \frac{7}{24} + 1 × (-1) × \frac{1}{3}\ + 1 × 0 × \frac{1}{6} + 1 × 1 × 0 =  -\frac{1}{2} $

2.Ковариационная матрица:

$ \text{Cov}(\xi, \eta) = \mathsf{E}(\xi \eta) - \mathsf{E}\xi \ \mathsf{E}\eta $

$ \text{Cov}(\xi, \eta) =  -\frac{1}{2} - 0 × (-\frac{1}{6}) =  -\frac{1}{2} $

Рассчитаем дисперсию:

$\mathbf{D}[\xi] = \mathsf{E}[\xi^2] - (\mathsf{E}[\xi])^2 = (-1)^2 \cdot \frac{1}{2} + 1^2 \cdot \frac{1}{2} - 0^2 = 1$

$\mathbf{D}[\eta] = \mathsf{E}[\eta^2] - (\mathsf{E}[\eta])^2 = (-1)^2 × \frac{11}{24}\ + 0^2 × \frac{1}{4} + 1^2 × \frac{7}{24} - (-\frac{1}{6})^2 = \frac{13}{18}$

Таким образом, ковариационная матрица будет иметь вид:

$
\Sigma(\xi, \eta) = \begin{pmatrix}
1 & -\frac{1}{2} \\
-\frac{1}{2} & \frac{13}{18}
\end{pmatrix}
$

3.Корреляционная матрица:

$ \rho(\xi, \eta) = \frac{\text{Cov}(\xi, \eta)}{\sqrt{\mathbf{D}\xi} \sqrt{\mathbf{D}\eta}} $

$\sqrt{\mathbf{D}\xi} = 1 $

$\sqrt{\mathbf{D}\eta} = \frac{\sqrt{26}}{6} $

$ \rho(\xi, \eta) = \frac{-\frac{1}{2}}{1 × \frac{\sqrt{26}}{6}} = -\frac{3\sqrt{26}}{26} $

Таким образом, корреляционная матрица будет иметь вид:

$
\Sigma(\xi, \eta) = \begin{pmatrix}
1 & -\frac{3\sqrt{26}}{26} \\
-\frac{3\sqrt{26}}{26} & 1
\end{pmatrix}
$

**(c) Исследовать $\xi$ и $\eta$ на независимость и некоррелированность**

Случайные величины независимы, если выполняется условие:

$$ P(\xi = x, \eta = y) = P(\xi = x) × P(\eta = y) $$ для всех $x$ и $y$.

Проверим, выполняется ли данное условие:

$ P(\xi = -1, \eta = 0) = \frac{1}{12} $

$ P(\xi = -1) = \frac{1}{2} $; $ P(\eta = 0) = \frac{1}{4} $

$ P(\xi = -1) × P(\eta = 0) = \frac{1}{2} × \frac{1}{4} \neq \frac{1}{12} $

Таким образом, делаем вывод, что случайные величины $\xi$ и $\eta$ являются **зависимыми**.

Случайные величины некоррелированы, если их ковариация равна 0:

$$ \text{Cov}(\xi, \eta) = 0 $$

В нашем случае ковариация равна:
$ \text{Cov}(\xi, \eta) = -\frac{1}{2} \neq {0} $

Таким образом, делаем вывод, что случайные величины $\xi$ и $\eta$ являются **коррелированными** (умеренная корреляция).

**Задание №5. Найти плотность распределения суммы двух независимых случайных величин $\xi$ и $\eta$, если $\xi \sim \text{Exp}(2)$ и $\eta \sim U(0,1)$.**

Случайная величина  $\xi$ имеет показательное (экспоненциальное) распределение с параметром $\lambda = 2$. Таким образом, плотность распределения $\xi$:

$ f_\xi(x) = \begin{cases}
0, & если \ x < 0 \\
2e^{-2x}, & если \ x \geq 0 \\
\end{cases}$

Случайная величина  $\eta$ имеет равномерное распределение на отрезке $[0, 1]$. Таким образом, плотность распределения $\eta$:

$ f_\eta(t) = \begin{cases}
0, & если \ \ t\notin [0, 1] \\
\frac{1}{b-a}= 1 & если \ t\in [0, 1]
\end{cases}$

Плотность распределения суммы случайных величин $\xi + \eta$ равна "свертке" плотностей $f_{\xi}$ и $f_{\eta}$:

$ f_{\xi + \eta}(t) = \int_{-\infty}^{\infty} f_{\xi}(x) f_{\eta}(t - x) \, dx $

Интеграл будет браться по области, где $x \geq 0$ и $0 \leq t - x \leq 1$ ($t-1 \leq x \leq t$), так как иначе будет 0.

$0 \leq t \leq 1$:

$ f_{\xi + \eta}(t) = \int_{0}^{t} (2e^{-2x} × 1 )dx = 2 \int_{0}^{t} e^{-2x}  dx = 2 \left[ -\frac{1}{2} e^{-2x} \right]_{0}^{t} = 1 - e^{-2t}$

$t > 1$:
   
$ f_{\xi + \eta}(t) = \int_{t-1}^{t} (2e^{-2x} × 1)  dx = 2 \int_{t-1}^{t} e^{-2x} dx = 2 \left[ -\frac{1}{2} e^{-2x} \right]_{t-1}^{t} = e^{-2(t-1)} - e^{-2t} = e^{2}e^{-2t} - e^{-2t} = e^{-2t}(e^{2} - 1)$

Таким образом, плотность распределения суммы случайных величин $\xi + \eta$ имеет вид:

$ f_{\xi + \eta}(t) = \begin{cases}
0,  & t < 0\\
1 - e^{-2t}, & 0 \leq t \leq 1 \\
e^{-2t}(e^{2} - 1), & t > 1
\end{cases} $
