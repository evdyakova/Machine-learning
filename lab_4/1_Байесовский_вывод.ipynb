# –ó–∞–¥–∞—á–∞ 1

–ü—É—Å—Ç—å $X_1, X_2, \ldots, X_n$ ‚Äî –≤—ã–±–æ—Ä–∫–∞ –∏–∑ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º $\lambda$. –ù–∞–π—Ç–∏ –æ—Ü–µ–Ω–∫—É –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\lambda$, —Å—Ä–∞–≤–Ω–∏—Ç—å –µ–µ —Å –±–∞–π–µ—Å–æ–≤—Å–∫–æ–π –æ—Ü–µ–Ω–∫–æ–π (MAP –∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∞–ø–æ—Å—Ç—Ä–µ–æ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è), –ø–æ–¥–æ–±—Ä–∞–≤ —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ. –°—Ä–∞–≤–Ω–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –±–∞–π–µ—Å–æ–≤—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ —Å –æ—Ü–µ–Ω–∫–æ–π MLE. –ù–∞–π—Ç–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

**–û—Ü–µ–Ω–∫–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è (MLE)**

–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ (—ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ) —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–º–µ–µ—Ç —Å–ª–µ–¥—É—é—â—É—é –ø–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

$
f_\xi(x) =
\begin{cases}
\lambda e^{-\lambda x}, & x \geqslant 0 \\
0, & \text{else}
\end{cases}
$

–§—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è:

$L = P(x_1) \cdot P(x_2) \cdot ... \cdot P(x_n) = Œªe^{-Œªx_1} \cdot Œªe^{-Œªx_2} \cdot ... \cdot Œªe^{-Œªx_n} = Œª^n e^{-Œª \sum_{i=1}^n X_i} $

–í–æ–∑—å–º–µ–º –ª–æ–≥–∞—Ä–∏—Ñ–º –æ—Ç $L$:

$ln(L) = ln(Œª^n e^{-Œª \sum_{i=1}^n X_i}) = n \ln \lambda - \lambda \sum_{i=1}^n X_i $

–ë–µ—Ä–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –ø–æ $Œª$:

$ \frac{\partial ln L}{\partial \lambda }  =  \frac{n}{\lambda} - \sum_{i=1}^n X_i = 0 $

$ \hat{\lambda} = \frac{1}{\bar{X}}$

**MAP –∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∞–ø–æ—Å—Ç—Ä–µ–æ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è**

–î–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $Œª$ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ–ø—Ä—è–∂–µ–Ω–Ω—ã–º —è–≤–ª—è–µ—Ç—Å—è –≥–∞–º–º–∞-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:

$Œª \sim \text{Gamma}(\alpha, \beta)$

–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –∞–ø—Ä–∏–æ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

$p(\lambda) = \frac{\beta^\alpha}{\Gamma(\alpha)} \lambda^{\alpha-1} e^{-\beta \lambda} $

–ê–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Ñ–æ—Ä–º—É–ª–µ –ë–∞–π–µ—Å–∞ –≤—ã–≥–ª—è–¥–∏—Ç —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$ p(\lambda | X) = \frac{p(X | \lambda) p(\lambda)}{P(X)} $

$ p(\lambda| X) = {\lambda^n e^{-\lambda \sum_{i=1}^n X_i}} \frac{\beta^\alpha}{\Gamma(\alpha)} \lambda^{\alpha-1} e^{-\beta \lambda} $

–£–ø—Ä–æ—â–∞–µ–º:

$ p(\lambda| X) = \lambda^{\alpha + n - 1} e^{-(\beta + \sum_{i=1}^n X_i) \lambda} $

–≠—Ç–æ –≥–∞–º–º–∞-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ $\alpha + n$ –∏ $\beta + \sum_{i=1}^n X_i$:

$ \textbf{P}(\lambda | X) \sim \text{Gamma}(\alpha + n, \beta + \sum_{i=1}^n X_i) $

MAP:

–î–ª—è –≥–∞–º–º–∞-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $ \hat{\lambda} = \frac{Œ± - 1}{Œ≤}, Œ± > 1$

$ \hat{\lambda} = \frac{Œ± + n - 1}{Œ≤ + \sum_{i=1}^n X_i}$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

$\mathbb{E}[\lambda | X] = \frac{\alpha + n}{\beta + \sum_{i=1}^n X_i} $

**–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –±–∞–π–µ—Å–æ–≤—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ —Å –æ—Ü–µ–Ω–∫–æ–π MLE**

MLE: $ \hat{\lambda} = \frac{1}{\bar{X}}$

MAP: $ \hat{\lambda} = \frac{Œ± + n - 1}{Œ≤ + \sum_{i=1}^n X_i}$

–ú–∞—Ç –æ–∂–∏–¥–∞–Ω–∏–µ: $\mathbb{E}[\lambda | X] = \frac{\alpha + n}{\beta + \sum_{i=1}^n X_i} $

–ú–æ–∂–Ω–æ –æ—Ç–º–µ—Ç–∏—Ç—å, —á—Ç–æ –æ—Ü–µ–Ω–∫–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç —Ç–æ–ª—å–∫–æ –¥–∞–Ω–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏ –∏ –ø–æ—Ç–æ–º—É –ø—Ä–∏ –º–∞–ª–æ–º n –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–æ–π –∏ —Å–º–µ—â—ë–Ω–Ω–æ–π. –ë–∞–π–µ—Å–æ–≤—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏  –≤–∫–ª—é—á–∞—é—Ç –∞–ø—Ä–∏–æ—Ä–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é, –ø–æ—ç—Ç–æ–º—É –æ–Ω–∏ –±–æ–ª–µ–µ —Å–≥–ª–∞–∂–µ–Ω–Ω—ã–µ.

–ü—Ä–∏ –º–∞–ª–æ–º —á–∏—Å–ª–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∑–∞–º–µ—Ç–Ω–æ –≤–ª–∏—è–Ω–∏–µ –∞–ø—Ä–∏–æ—Ä–∞, –∞ –ø—Ä–∏ —Ä–æ—Å—Ç–µ –æ–±—ä—ë–º–∞ –≤—ã–±–æ—Ä–∫–∏ –≤—Å–µ –æ—Ü–µ–Ω–∫–∏ —Å–±–ª–∏–∂–∞—é—Ç—Å—è –∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏ —Å–æ–≤–ø–∞–¥–∞—é—Ç.

**–ü—Ä–µ–¥—Å–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ**

–ü—Ä–µ–¥—Å–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª—è –Ω–æ–≤–æ–≥–æ $X_{n+1}$, –∏—Å–ø–æ–ª—å–∑—É—è –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:

$ p(X_{n+1} | X_n) = \int_0^\infty p(X_{n+1} | \lambda) p(\lambda | X) d\lambda $

–ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:

$ p(X_{n+1} | X_n) = \int_0^\infty \lambda e^{-\lambda X_{n+1}} \frac{(\beta + \sum_{i=1}^n x_i)^{\alpha + n}}{\Gamma(\alpha + n)} \lambda^{\alpha + n - 1} e^{-(\beta + \sum_{i=1}^n X_i) \lambda}dŒª = \frac{(\beta + \sum_{i=1}^n X_i)^{\alpha + n}}{\Gamma(\alpha + n)}\int_0^\infty \lambda^{\alpha + n} e^{-\lambda (X_{n+1} + \beta + \sum_{i=1}^n X_i)} d\lambda $

–ò—Å–ø–æ–ª—å–∑—É–µ–º —Å–≤–æ–π—Å—Ç–≤–æ –≥–∞–º–º–∞-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\Gamma(\alpha_n + 1) = \alpha_n \Gamma (\alpha_n) $:

$ p(X_{n+1} | X_n) = \frac{(\alpha + n) (\beta + \sum_{i=1}^n X_i)^{\alpha + n}}{(X_{n+1} + \beta + \sum_{i=1}^n X_i)^{\alpha + n + 1}}, X_{n+1}  \geqslant 0 $


# –ó–∞–¥–∞—á–∞ 2

**–ú—É–ª—å—Ç–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ**

–ü—É—Å—Ç—å –ø—Ä–æ–≤–æ–¥–∏—Ç—Å—è —Å–µ—Ä–∏—è –∏–∑ $n$ –∏—Å–ø—ã—Ç–∞–Ω–∏–π –∏ –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –∫–∞–∂–¥–æ–≥–æ –∏—Å–ø—ã—Ç–∞–Ω–∏—è –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç —Ä–æ–≤–Ω–æ –æ–¥–Ω–æ —Å–æ–±—ã—Ç–∏–µ –∏–∑ –Ω–∞–±–æ—Ä–∞ $A_1, A_2, \dots, A_m$, –ø—Ä–∏—á–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —ç—Ç–∏—Ö —Å–æ–±—ã—Ç–∏–π —Ä–∞–≤–Ω—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ $\mathsf{p}_1, \mathsf{p}_2, \dots, \mathsf{p}_m$, –ø—Ä–∏—á–µ–º
$$\sum_{i=1}^{m}\mathsf{p}_i = 1.$$

–¢–æ–≥–¥–∞ —Å–æ–≤–º–µ—Å—Ç–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ–ª–∏—á–∏–Ω $X_1, X_2, \dots, X_m$, –≥–¥–µ $X_k$ ‚Äî —á–∏—Å–ª–æ –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏–π —Å–æ–±—ã—Ç–∏—è $A_k$ –≤ —Å–µ—Ä–∏–∏ –∏–∑ $n$ –∏—Å–ø—ã—Ç–∞–Ω–∏–π, –∑–∞–¥–∞–µ—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏

$$
\mathsf{P}\left(X_1 = n_1, \dots, X_m = n_m, \right) = \frac{n!}{n_1!\dots n_m!}\mathsf{p}_1^{n_1}\dots \mathsf{p}_m^{n_m},
$$

–≥–¥–µ $n_1, n_2, \dots, n_m$ ‚Äî¬†–ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–π –Ω–∞–±–æ—Ä —Ü–µ–ª—ã—Ö –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, —Ç–∞–∫–∏—Ö —á—Ç–æ

$$\sum_{i=1}^m n_i = n.$$

–ü—Ä–æ–∏–∑–≤–µ–¥–∏—Ç–µ –±–∞–π–µ—Å–æ–≤—Å–∫–∏–π –≤—ã–≤–æ–¥ –¥–ª—è –º—É–ª—å—Ç–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: –Ω–∞–π–¥–∏—Ç–µ –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, –∏—Å–ø–æ–ª—å–∑—É—è –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Å–æ–ø–æ—Ä—è–∂–µ–Ω–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—é [—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –î–∏—Ä–∏—Ö–ª–µ](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%94%D0%B8%D1%80%D0%B8%D1%85%D0%BB%D0%B5), –Ω–∞–π–¥–∏—Ç–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ. –û–±—ä—è—Å–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

–§—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è –±—É–¥–µ—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$L(\mathsf{p}|X) = \frac{n!}{n_1!\dots n_m!}\mathsf{p}_1^{n_1}\dots \mathsf{p}_m^{n_m}$

–î–ª—è –º—É–ª—å—Ç–∏–Ω–æ–º–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ–ø—Ä—è–∂–µ–Ω–Ω—ã–º –±—É–¥–µ—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –î–∏—Ä–∏—Ö–ª–µ.

$ \textbf{p}\sim \text{Dir}(\alpha_1, ... \alpha_m) $

–§—É–Ω–∫—Ü–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –≤—ã–≥–ª—è–¥–∏—Ç —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$ p(\mathsf{p}_1, \mathsf{p}_2, \dots, \mathsf{p}_m) = \frac{\Gamma(\sum_{i=1}^m \alpha_i)}{\prod_{i=1}^m \Gamma(\alpha_i)} \prod_{i=1}^m \mathsf{p}_i^{\alpha_i - 1}$

**–ê–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ**

–ü–æ —Ñ–æ—Ä–º—É–ª–µ –ë–∞–π–µ—Å–∞:

$p(\mathbf p| X) \propto p(X | \mathbf p)\,p(\mathbf p) $

$p(\mathbf p| X) \propto  \prod_{i=1}^m \mathsf{p}_i^{n_i} \cdot \prod_{i=1}^m \mathsf{p}_i^{\alpha_i - 1} = \prod_{i=1}^m \mathsf{p}_i^{\alpha_i + n_i - 1}$

MAP:

$ \hat{p_i} = \frac{Œ±_i + n_i - 1}{\sum_{j=1}^{m} (Œ±_j + n_j) -m}$, –ø—Ä–∏ $Œ±_i + n_i > 1$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ:

$\mathbb{E}[p_i | X] = \frac{\alpha_i + n_i}{\sum_{j=1}^m (\alpha_j + n_j)} $

–ú–æ–∂–Ω–æ –æ—Ç–º–µ—Ç–∏—Ç—å, —á—Ç–æ –±–∞–π–µ—Å–æ–≤—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ —É—á–∏—Ç—ã–≤–∞—é—Ç –∫–∞–∫ –¥–∞–Ω–Ω—ã–µ, —Ç–∞–∫ –∏ –∞–ø—Ä–∏–æ—Ä, —á—Ç–æ –¥–∞–µ—Ç —Å–≥–ª–∞–∂–µ–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏. –ü—Ä–∏ –º–∞–ª–æ–º $n$ –≤–ª–∏—è–Ω–∏–µ –∞–ø—Ä–∏–æ—Ä–∞ –∑–∞–º–µ—Ç–Ω–æ, —Å —Ä–æ—Å—Ç–æ–º –∂–µ –æ—Ü–µ–Ω–∫–∏ —Å—Ç—Ä–µ–º—è—Ç—Å—è –∫ —á–∞—Å—Ç–æ—Ç–Ω—ã–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º.

**–ü—Ä–µ–¥—Å–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ**

–î–ª—è –º—É–ª—å—Ç–∏–Ω–æ–º–∏–Ω–∞–ª—å–Ω–æ–π –º–æ–¥–µ–ª–∏:

$p (X_{n+1} = A_k|X) = ùîº[p_k|X] = \frac{\alpha_k + n_k}{\sum_{i=1}^m(\alpha_i + n_i)}$
